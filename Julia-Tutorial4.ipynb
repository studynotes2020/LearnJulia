## Data Visualization


```julia
using Plots
```

```julia
x = 1:15; y = rand(15);
```

```julia
plot(x,y)
```

```julia
z = rand(15);
plot!(x, z)
```

```julia
plot(x, y, title="One random chart", label = "Line Chart 1")
```

```julia
plot(x,y,xlabel="X axis label", ylabel="Y axis Label")
```

```julia
scatter(x,y)
```

```julia
bar(x,y)
```

```julia
histogram(y)
```

```julia
p1 = plot(x,y)
p2 = scatter(x,y)
p3 = bar(x,y)
p4 = histogram(y)

plot(p1, p2, p3, p4, layout=(2,2), legend=false)
```

```julia
y = rand(15,4)
plot(x,y,layout=(4,1))
```
